```python
# Assuming df is your DataFrame containing the text data
# Defining a Function to Pull Words
def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)

# Toknizing the Given Text
testtext = "The Abbott and Costello Cartoon Show"
testtokens = tokenize(testtext)
print(testtokens)
```

```python
# Importing nltk Package
import nltk

# Downloading all List of Stop Words
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)
```

```python
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords] # Define a function 'remove_stop' to remove stopwords from a list of tokens

cleantokens = remove_stop(testtokens)  # Test the 'remove_stop' function on cleaned tokens
print(cleantokens) # Displaying cleantokens
```

```python
# Creating a Pipeline
pipeline = [str.lower, tokenize, remove_stop]

# Defining a Function to Run the Pipeline
def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

```python
# Running the Pipeline on DataFrame
df['tokens'] = df['American company'].apply(prepare, pipeline=pipeline)
```

```python
#creating a New Column for the Number of Tokens
df['num_tokens'] = df['tokens'].map(len)
df.tail()
```

```python
# Function to remove emoji and Unicode characters from text
def remove_emoji(string):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)

# Cleaning the text in the 'Titles' column using the remove_emoji function
df["Title"]= df["Title"].apply(remove_emoji)

# Displaying the clean text
df.head()
```

```python
# Function to get the subjectivity of text using TextBlob
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

# Function to get the polarity of text using TextBlob
def getPolarity(text):
    return TextBlob(text).sentiment.polarity

# Adding new columns for subjectivity and polarity to the bitcoin_df DataFrame
df['Subjectivity'] = df['Title'].apply(getSubjectivity)
df['Polarity'] = df['Title'].apply(getPolarity)

# Displaying the updated DataFrame
df.head()
```

```python
# Function to categorize polarity scores into negative, neutral, or positive categories
def getInsight(score):
    if score < 0:
        return "Negative"
    elif score == 0:
        return "Neutral"
    else:
        return "Positive"
    
# Applying the getInsight function to create a new column 'Insight' in the bitcoin_df DataFrame
df["Insight"] = df["Polarity"].apply(getInsight)

# Displaying the first 50 rows of the DataFrame
df.head(50)
```

```python
import nltk
from nltk.tokenize import word_tokenize

# Downloading the opinion lexicon dataset
nltk.download('opinion_lexicon')

# Importing opinion_lexicon from nltk.corpus
from nltk.corpus import opinion_lexicon

# Define positive and negative scores
pos_score = 1
neg_score = -1

# Initialize a dictionary to store words and their scores
word_dict = {}

# Adding positive words to the dictionary with positive score
for word in opinion_lexicon.positive():
    word_dict[word] = pos_score

# Adding negative words to the dictionary with negative score
for word in opinion_lexicon.negative():
    word_dict[word] = neg_score

# Function to calculate Bing Liu score for a given text
def bing_liu_score(text):
    sentiment_score = 0  # Initialize sentiment score
    # Tokenize the text into words
    words = word_tokenize(text.lower())
    # Iterate over each word and calculate the sentiment score
    for word in words:
        if word in word_dict:
            sentiment_score += word_dict[word]
    return sentiment_score

# Apply the function to the 'Title' column to create a new column 'Bing_Liu_Score'
df['Bing_Liu_Score'] = df['Title'].apply(bing_liu_score)

# Display sample of DataFrame with Bing Liu scores
print(df[['length', 'Title', 'Bing_Liu_Score']].sample(10))
```

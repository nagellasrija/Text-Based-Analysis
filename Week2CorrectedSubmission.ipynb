# Tokenization Function using Regex

```python
# Importing Regex Package
import regex as re
# Defining a Function to Pull Words
def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)
```

# Testing the 'tokenize' Function

```python
# Toknizing the Given Text
testtext = "Let's defeat SARS-CoV-2 together in 2020! www.cdc.gov"
testtokens = tokenize(testtext)
print(testtokens)
```

# Downloading and Using NLTK Stopwords

```python
# Importing nltk Package
import nltk
# Downloading all List of Stop Words
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)
```

# Removing Stopwords from Tokens

```python
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords] # Define a function 'remove_stop' to remove stopwords from a list of tokens
cleantokens= remove_stop(testtokens)  # Test the 'remove_stop' function on cleaned tokens
print(cleantokens) #Displaying cleantokens
```

# Text Preparation Pipeline

```python
# Creating a Pipeline
pipeline = [str.lower, tokenize, remove_stop]

# Defining a Function to Run the Pipeline
def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

# Applying Text Preparation Pipeline to Create 'tokens' Column

```python
# Running the Pipeline
df['tokens'] = df['text'].apply(prepare, pipeline=pipeline)
```

# Applying Text Preparation and Tokenization to Create 'tokens' and 'num_tokens' Columns

```python
#creating a New Column for the Number of Tokens
df['num_tokens'] = df['tokens'].map(len)
df.tail()
```

# Testing Counter Class on a Sample List of Tokens

```python
# importing a function Counter
from collections import Counter

tokens = tokenize("She likes my cats and my cats like my sofa.")
testcount= Counter(tokens)
print(testcount)
```

# Updating Counter with More Tokens

```python
# Addition of New Data to the Current Count
more_tokens = tokenize("She likes dogs and cats.")
testcount.update(more_tokens)
print(testcount)
```

# Updating Counter with More Tokens

```python
# Printing the 5 Words Repeated the Most
counter = Counter()
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

# Creating DataFrame 'freq_df' from Counter and Filtering Tokens

```python
# Create a DataFrame 'freq_df' from the Counter, filter tokens with frequency greater than or equal to 1
freq_df = pd.DataFrame.from_dict(counter,orient='index', columns=['freq']) 
freq_df = freq_df.query('freq >= 1') #Filtering Tokens with Frequency >= 1
freq_df.index.name = 'token' #Setting the Index Name
```

# Sorting and Displaying Top Tokens by Frequency

```python
freq_df = freq_df.sort_values('freq', ascending=False) #Sorting 'freq_df' by Frequency
freq_df.head(10)  # Display the top 10 tokens based on frequency
```
